```python
import cv2
import numpy as np 
import pandas as pd
from keras.models import load_model
model=load_model(r"C:\Users\Yash Jain\Desktop\CVIP\Pneumonina_CNN.h5")
```

```python
import glob
cv_img =[]
for img in glob.glob(r"C:\Users\Yash Jain\Desktop\chest_xray\test\NORMAL\*.jpeg"):
    n=cv2.imread(img)
    cv_img.append(n)
```

```python
type(cv_img)
```

```python
cv_img=np.array(cv_img)
cv_img[3].shape
```

```python
prediction_class=np.zeros((234,1))
for i in range(234):
    cv_img[i] =cv2.resize(cv_img[i],(64,64), interpolation = cv2.INTER_AREA)
    cv_img[i]=np.expand_dims(cv_img[i],axis=0)
    if(np.max(cv_img[i])>1):
            cv_img[i]=cv_img[i]/255.0
    prediction_class[i]=model.predict_classes(cv_img[i])
    print(prediction_class)
```

```python
count=0
count1=0
for i in prediction_class:
    if(i==0):
        count=count+1
    if(i==1):
        count1=count1+1
print(count)
print(count1)
```
